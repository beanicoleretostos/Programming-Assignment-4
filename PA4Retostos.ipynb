```python
# Step 2: Load the csv file
import pandas as pd


df = pd.read_excel('board2.xlsx')
df
```

```python
#### Students from Luzon

# Step 1: Filter DataFrame based on conditions
instru = df[(df['Track'] == 'Instrumentation') & (df['Hometown'] == 'Luzon') & (df['Electronics'] > 70)].copy()

# Step 2: Select specific columns
instru = instru[['Name', 'GEAS', 'Electronics']]
```

```python
#### Female studedntrs from Mindanao

# Step 1: Filter DataFrame for Mindanao females (using .query) 
mindy = df.query('Hometown == "Mindanao" and Gender == "Female"').copy() 

# Step 2: Calculate the average of selected columns and add it as a new column
mindy['Average'] = mindy[['Math', 'GEAS', 'Electronics']].mean(axis=1) 

# Step 3: Filter the DataFrame to include only rows where the average is 55 or higher
mindy = mindy[mindy['Average'] >= 55]
```

```python
# Select specific columns using .loc
mindy = mindy.loc[:, ['Name', 'Track', 'Electronics', 'Average']]
```

```python
import matplotlib.pyplot as plt
import pandas as pd

#### Prepare the data (assuming 'df' is already defined and includes 'Hometown' and 'Average')
plt.figure(figsize=(10, 6))

# Create a boxplot using matplotlib
df.boxplot(column='Average', by='Hometown', grid=False)

# Add titles and labels
plt.title('Average Scores by Hometown')
plt.suptitle('')  # Remove the default title to keep the plot clean
plt.xlabel('Hometown')
plt.ylabel('Average')

# Show the plot
plt.show()
```
